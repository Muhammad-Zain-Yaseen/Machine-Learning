```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering 
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
#silhioutte_score will tell you how well the data has been clustered. it ranges from -1 to 1.
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/My Drive/Datasets/bank.txt', header = None, names=['variance','skewness','curtosis', 'entropy', 'class'])
data
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.drop('class', axis=1))
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(data_scaled)
predictions = kmeans.predict(data_scaled)
```

```python
# Visualize the results
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=predictions)
plt.show()
```

```python
# Calculate the silhouette score
score = silhouette_score(data_scaled, predictions)
print(f'Silhouette score: {score:.3f}')
```

```python
# Apply agglomerative clustering
agg = AgglomerativeClustering(n_clusters=2)
predictions = agg.fit_predict(data_scaled)
```

```python
# Visualize the results
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=predictions)
plt.show()
```

```python
# Calculate the silhouette score
score = silhouette_score(data_scaled, predictions)
print(f'Silhouette score: {score:.3f}')
```
